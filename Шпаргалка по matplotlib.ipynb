```python
import matplotlib.pyplot as plt
import numpy as np
```

Создадим данные для работы

```python
np.linspace(0,9,10) # функция linspace  разбивает промежуток от 0 до 9 на 10 частей
```

```python
x = np.linspace(1,10,1000)
y1 = np.power(x, 0.5)
y2 = 2*np.sin(x)
y3 = 3*np.cos(x)
```

```python
plt.plot(x, y1)
```

```python
#чтобы расположить несколько графиков на одной плоскоски 
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
# добавим оси к графикам
plt.plot(x, y1)
plt.plot(x, y2)

plt.xlabel('x') # Ось х
plt.ylabel('f(x)') # Ось у
plt.title('График 1') # Заголовок
```

```python
# чтобы на графике обозначить название каждой криво, нужно воспользоваться параметром label
plt.plot(x, y1, label = 'f1')
plt.plot(x, y2, label = 'f2')

plt.legend() # функция, чтобы напечатать название кривых f1 и f2
plt.xlabel('x') # Ось х
plt.ylabel('f(x)') # Ось у
plt.title('График 1') # Заголовок
```

```python
# размеры гарфика
plt.plot(x, y1, label = 'f1')
plt.plot(x, y2, label = 'f2')

plt.legend() # функция, чтобы напечатать название кривых f1 и f2
plt.xlabel('x') # Ось х
plt.ylabel('f(x)') # Ось у
plt.title('График 1') # Заголовок

plt.xlim(1,5)
```

```python
# замена цвета и формата линии
plt.plot(x, y1, label = 'f1')
plt.plot(x, y2, label = 'f2')
plt.plot(x, y3, 'b--', label = 'f3') # b-- означает синий цвет и пунктирная линия

plt.legend() # функция, чтобы напечатать название кривых f1 и f2
plt.xlabel('x') # Ось х
plt.ylabel('f(x)') # Ось у
plt.title('График 1') # Заголовок
```

```python
# два графика на одной странице
fig, axes = plt.subplots(nrows = 1, ncols = 2) # объект fig-стиль отображения гр-ов, axes - области на которых расположены гр-и 
ax1, ax2 = axes # присваиваем переменную каждому окну 
# вызываем гарфики
ax1.plot(x,y1)
ax2.plot(x,y2)
```

```python
# изменение размера областей 

fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (10,2), dpi = 300) # figsize параметр размера окна графиков, dpi-разр-ие на дюйм
ax1, ax2 = axes # присваиваем переменную каждому окну 
# вызываем гарфики
ax1.plot(x,y1)
ax1.set_xlabel('x') # подписать ось х на первом графике
ax2.plot(x,y2)
```

```python
#Чтобы редактировать иощий объект, требуется использовать объект figo
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (10,2), dpi = 300) # figsize параметр размера окна графиков, dpi-разр-ие на дюйм
ax1, ax2 = axes # присваиваем переменную каждому окну 

ax1.plot(x,y1)
ax1.set_xlabel('x')
ax2.plot(x,y2)
ax2.set_xlabel('x')

fig.suptitle('Графики')
```

ТИПЫ ДАННЫХ В matplotlib

```python
names = ['a', 'b', 'c']
values = [100, 150, 180]

fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 10), dpi = 300)

# т к axes массив 2 на 2, нужно создавать переменные таким отбразом:
(ax1, ax2), (ax3, ax4) = axes

ax1.bar(names, values, color = ['r', 'g', 'b'])
ax1.set_title('bar')
ax2.scatter(names, values)
ax2.set_title('scatter')
ax3.plot(names, values)
ax3.set_title('plot')
ax4.pie( values)
ax4.set_title('pie')
```
